# Filtering dataframe on a condition | Masking using boolean series

```python
data.head()
```

```python
data[data['city'] == 'Hyderabad']
```

```python
#data[data['city'] == 'Hyderabad'].shape
data[data['city'] == 'Hyderabad'].shape[0]
```

```python
def get_city_match_count(city):
    return data[data['city'] == city].shape[0]
```

```python
#get_city_match_count('Mumbai')
get_city_match_count('Bangalore')
```

```python
mask1 = data['city'] == 'Hyderabad'
mask2 = data['date'] > '2010-01-01'

data[mask1 & mask2]    # so whenever we do and condintion so we apply this (& operator) if both side are in boolean. otherwise
                        # normal and.
```

```python
data[(data['city'] == 'Hyderabad') & (data['date'] > '2010-01-01')]  # this way you have to give parenthesis in both Eqn.
```
